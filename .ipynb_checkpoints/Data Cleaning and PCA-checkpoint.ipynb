```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *

# define a function for calculating the metric to be used later 
def classification_metrics(Y_pred, Y_true):
    acc = accuracy_score(Y_true, Y_pred)
    precision = precision_score(Y_true, Y_pred)
    recall = recall_score(Y_true, Y_pred)
    f1score = f1_score(Y_true, Y_pred)
    auc = roc_auc_score(Y_true, Y_pred)

    return acc, precision, recall, f1score, auc

# define a function for printing the metrics 
def display_metrics(classifierName,Y_pred,Y_true):
    print ("_____________________________")
    print ("Model: "+classifierName)
    acc, precision, recall, f1score, auc = classification_metrics(Y_pred,Y_true)
    print ("Accuracy: "+str(acc))
    print ("Precision: "+str(precision))
    print ("Recall: "+str(recall))
    print ("F1-score: "+str(f1score))
    print ("AUC: "+str(auc))
    print ("_____________________________")
    print ("")
```

```python
# Load the dataset
data=pd.read_csv('/Users/von/Documents/SPU/202309 AUTUMN/ISM6353/Project/satisfaction.csv')
data
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Remove missing values
data_clean = data.dropna()
data_clean = data_clean.drop('id',axis=1)
data_clean = pd.get_dummies(data_clean, columns=['Gender','Customer Type','Type of Travel','Class'],drop_first=True)
data_clean
```

```python
# Set target variable and the features
target = 'satisfaction_v2'
y = data_clean[target]
X = data_clean.drop(columns=[target])
```

```python
# Convert target variable as categorical
y = pd.get_dummies(y, drop_first=True).reset_index(drop=True)
y
```

```python
# Standardize the data
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)

features=pd.DataFrame(X_scaled)
features
data_pca=pd.concat([y,features],axis=1)
data_pca
```

```python
# Initialize PCA
#pca = PCA()

# Fit PCA on the scaled data
#pca.fit(X_scaled)

pca = PCA(random_state = 101).fit(X_scaled) 

components = list(range(1, X_scaled.shape[1]+1)) # get a list of # of components
cumulative_var = np.cumsum(pca.explained_variance_ratio_) # use np.cumsum function to calculate cumulative var by # of components

plt.plot(components, cumulative_var)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance');


# Now we can look at the explained variance to determine how many components to keep
explained_variance = pca.explained_variance_ratio_

# Display the cumulative sum of the explained variance
cumulative_explained_variance = explained_variance.cumsum()

cumulative_explained_variance
```

```python
# Choosing the number of components to keep
n_components = np.where(cumulative_explained_variance > 0.9)[0][0] + 1  # plus one because index starts at 0

# Re-fit PCA with the selected number of components
pca_selected = PCA(n_components=n_components)
X_pca = pca_selected.fit_transform(X_scaled)

# Get the loadings of the first few components
loadings = pca_selected.components_.T * np.sqrt(pca_selected.explained_variance_)

# Create a DataFrame of the loadings with the original feature names
loadings_df = pd.DataFrame(loadings, columns=[f'PC{i+1}' for i in range(n_components)], index=X.select_dtypes(include=[np.number]).columns)

# Display the loadings for the principal components
loadings_df.head(n_components)  # Show loadings for the number of components we kept
```

```python

# Create a heatmap for the loadings of the top components
plt.figure(figsize=(10, 10))
sns.heatmap(loadings_df.iloc[:, :n_components], cmap='YlGnBu', annot=True, fmt=".2f")
plt.title('Absolute Loadings of Variables on Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Variables')
plt.show()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=101)

# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate the accuracy and generate a classification report
accuracy = accuracy_score(y_test, y_pred)

print('Acuracy:', accuracy) 
display_metrics('Logistic',y_test, y_pred)
```

```python
accuracies = [] # compare accuracies for using diff # of components
components = list(range(1, n_components+1)) # a list of 1 to 13

for i in range(1, n_components+1): # loop through 1 to 13 
    
    # define PCA
    pca = PCA(n_components = i, random_state = 101) 
    
    # fit PCA to data
    X_pca = pca.fit_transform(X_scaled)
        
    # split data into train/test data
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state = 101)
    
    # use solver = 'lbfgs', multi_class = 'multinomial' b/c we have multiple categories for target var
    model_LR = LogisticRegression(solver = 'lbfgs', multi_class = 'multinomial', max_iter = 5000)

    model_LR.fit(X_train, y_train)
    
    y_pred = model_LR.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)

    accuracies.append(acc)
    
    print('components:', i, ', accuracy:', acc)
    



```

```python
plt.plot(components, accuracies)
plt.ylabel('Accuracy')
plt.xlabel('# of Principal Components')
plt.title('Accuracy vs. # of Principal Components')
```

```python
#Selecting number of component which return the highest accuracy
model_LR_results=pd.DataFrame({'Components':components,'Accuracy':accuracies}).sort_values(by='Accuracy',ascending=False).reset_index(drop=True)
n_components_selected=model_LR_results.iloc[0,0]
model_LR_results
```

```python
# Re-fit PCA with the selected number of component
pca = PCA(random_state=101)
X_pca = pca.fit_transform(X_scaled)
pca_selected = PCA(n_components_selected, random_state=101)
X_pca_selected = pca_selected.fit_transform(X_scaled)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca_selected, y, test_size=0.3, random_state=101)

# Initialize and train the logistic regression model
model_selected = LogisticRegression(max_iter=5000, random_state=101)
model_selected.fit(X_train, y_train.values.ravel())

#y_pred_test = model_selected.predict(X_test)
    
#acc_test = accuracy_score(y_test, y_pred)
#acc_test

# Get the model coefficients (for the last trained model, which has the highest number of components)
coefficients = model_selected.coef_[0]

# Compute feature importance for the original features
# Multiply the loadings by the coefficients
original_feature_importance = pca_selected.components_.T @ coefficients

# Create a DataFrame for the feature importance
feature_importance_df = pd.DataFrame(original_feature_importance, index=X.columns, columns=['Importance'])

# Sort the features by importance
feature_importance_sorted = feature_importance_df.abs().sort_values(by='Importance', ascending=False)

feature_importance_sorted.head() 

plt.figure(figsize=(10, 10))
feature_importance_sorted.plot(kind='barh', legend=None, color='cornflowerblue')
plt.title('Feature Importance for Logistic Regression Model')
plt.ylabel('Importance Score')
plt.xlabel('Features')
plt.gca().invert_yaxis()
plt.show

display_metrics('Logistic with PCA',y_test, y_pred)
```
